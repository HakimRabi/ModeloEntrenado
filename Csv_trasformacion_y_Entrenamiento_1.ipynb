```python
import py7zr
import os

def descomprimir_7z(archivo_7z, carpeta_destino):
    # Verificar si la carpeta de destino existe, si no, crearla
    if not os.path.exists(carpeta_destino):
        os.makedirs(carpeta_destino)

    with py7zr.SevenZipFile(archivo_7z, 'r') as zip_ref:
        zip_ref.extractall(carpeta_destino)
        print(f"¡{archivo_7z} ha sido descomprimido correctamente en {carpeta_destino}!")

if __name__ == "__main__":
    # Ruta del archivo .7z que quieres descomprimir
    archivo_7z = "/content/T28052_CNANCUAN_2017-2023.7z"
    # Ruta de la carpeta donde quieres almacenar los archivos descomprimidos
    carpeta_destino = "/content/Datasets_todos"

    # Llamar a la función para descomprimir el archivo .7z
    descomprimir_7z(archivo_7z, carpeta_destino)
```

```python
import pandas as pd

def estudio_completitud(nombre_archivo):
    # Leer el archivo CSV en un DataFrame
    df = pd.read_csv(nombre_archivo)

    # Contar los valores faltantes en cada columna
    faltantes_por_columna = df.isnull().sum()

    # Calcular el total de filas
    total_filas = len(df)

    # Calcular el porcentaje de completitud para cada columna
    print("Estudio de Completitud:")
    for columna, faltantes in faltantes_por_columna.items():
        completitud = ((total_filas - faltantes) / total_filas) * 100
        print(f"La columna '{columna}' tiene {faltantes} valores faltantes, completitud: {completitud:.2f}%")

if __name__ == "__main__":
    # Nombre del archivo CSV que quieres analizar
    nombre_archivo = "/content/archivo_final_dataset.csv"

    # Realizar el estudio de completitud
    estudio_completitud(nombre_archivo)
```

```python
import pandas as pd

def estadisticas_descriptivas(nombre_archivo):
    # Leer el archivo CSV en un DataFrame
    df = pd.read_csv(nombre_archivo)

    # Calcular estadísticas descriptivas adicionales
    estadisticas = {
        'count': df.count(),
        'missing': df.isnull().sum(),
        'mean': df.mean(),
        'median': df.median(),
        'std': df.std(),
        'min': df.min(),
        'max': df.max(),
        '25th_percentile': df.quantile(0.25),
        '50th_percentile': df.quantile(0.50),
        '75th_percentile': df.quantile(0.75),
        'skewness': df.skew(),
        'kurtosis': df.kurtosis()
    }

    # Imprimir las estadísticas descriptivas
    print("Estadísticas Descriptivas:")
    for columna, valores in estadisticas.items():
        print(f"\nColumna: {columna}")
        print(valores)

if __name__ == "__main__":
    # Nombre del archivo CSV que quieres analizar
    nombre_archivo = "/content/archivo_final_dataset.csv"

    # Generar estadísticas descriptivas
    estadisticas_descriptivas(nombre_archivo)

```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import mean_squared_error, accuracy_score

# Cargar el conjunto de datos desde el archivo CSV
df = pd.read_csv("/content/nuevo_archivo_final_dataset_con_subestacion.csv")

# Separar las características (X) y las etiquetas (y) para la predicción del consumo
X_consumo = df[['hora']]
y_consumo = df['consumo']

# Separar las características (X) y las etiquetas (y) para la predicción de la subestación
X_subestacion = df[['hora']]
y_subestacion = df['sub_estacion']

# Dividir los datos en conjuntos de entrenamiento y prueba para ambas predicciones
X_consumo_train, X_consumo_test, y_consumo_train, y_consumo_test = train_test_split(X_consumo, y_consumo, test_size=0.2, random_state=42)
X_subestacion_train, X_subestacion_test, y_subestacion_train, y_subestacion_test = train_test_split(X_subestacion, y_subestacion, test_size=0.2, random_state=42)

# Inicializar y entrenar el modelo de regresión lineal para la predicción del consumo
modelo_consumo = LinearRegression()
modelo_consumo.fit(X_consumo_train, y_consumo_train)

# Predecir el consumo en el conjunto de prueba y calcular la precisión
predicciones_consumo = modelo_consumo.predict(X_consumo_test)
precision_consumo = mean_squared_error(y_consumo_test, predicciones_consumo)
print("Precisión del modelo de regresión lineal para la predicción del consumo:", precision_consumo)

# Inicializar y entrenar el modelo KNN para la predicción de la subestación
modelo_subestacion = KNeighborsClassifier()
modelo_subestacion.fit(X_subestacion_train, y_subestacion_train)

# Predecir la subestación en el conjunto de prueba y calcular la precisión
predicciones_subestacion = modelo_subestacion.predict(X_subestacion_test)
precision_subestacion = accuracy_score(y_subestacion_test, predicciones_subestacion)
print("Precisión del modelo KNN para la predicción de la subestación:", precision_subestacion)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN
from sklearn.metrics import mean_squared_error

# Cargar el conjunto de datos desde el archivo CSV
df = pd.read_csv("/content/nuevo_archivo_final_dataset_con_subestacion.csv")

# Preparar los datos para el modelo RNN
X_consumo_rnn = df[['hora']].values
y_consumo_rnn = df['consumo'].values

# Escalar los datos
scaler = MinMaxScaler(feature_range=(0, 1))
X_consumo_rnn_scaled = scaler.fit_transform(X_consumo_rnn)
y_consumo_rnn_scaled = scaler.fit_transform(y_consumo_rnn.reshape(-1, 1))

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train_rnn, X_test_rnn, y_train_rnn, y_test_rnn = train_test_split(X_consumo_rnn_scaled, y_consumo_rnn_scaled, test_size=0.2, random_state=42)

# Definir la arquitectura del modelo RNN
modelo_rnn = Sequential([
    SimpleRNN(50, activation='relu', input_shape=(X_train_rnn.shape[1], 1)),
    Dense(1)
])

# Compilar el modelo RNN
modelo_rnn.compile(optimizer='adam', loss='mean_squared_error')

# Entrenar el modelo RNN
modelo_rnn.fit(X_train_rnn.reshape((X_train_rnn.shape[0], X_train_rnn.shape[1], 1)), y_train_rnn, epochs=100, batch_size=32, verbose=1)

# Realizar predicciones en el conjunto de prueba
predicciones_rnn = modelo_rnn.predict(X_test_rnn.reshape((X_test_rnn.shape[0], X_test_rnn.shape[1], 1)))

# Escalar las predicciones inversamente
predicciones_rnn_inverso = scaler.inverse_transform(predicciones_rnn)

# Calcular la precisión del modelo RNN
precision_rnn = mean_squared_error(y_test_rnn, predicciones_rnn)
print("Precisión del modelo RNN para la predicción del consumo:", precision_rnn)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.callbacks import ModelCheckpoint

# Cargar datos
data = pd.read_csv('nuevo_archivo_final_dataset_con_subestacion.csv')

# Seleccionar una subestación específica
sub_estacion = 'ELSALTO'  # Reemplaza esto con el nombre de la subestación que quieras seleccionar
data_subestacion = data[data['sub_estacion'] == sub_estacion]

# Seleccionar columnas relevantes y normalizar datos
scaler = MinMaxScaler()
data_normalized = scaler.fit_transform(data_subestacion[['consumo', 'dia', 'mes', 'año', 'hora']])

# Preparar datos para RNN
X = []
y = []
sequence_length = 24  # Longitud de la secuencia de entrada (por ejemplo, usar datos de las últimas 24 horas para predecir el siguiente consumo)

for i in range(len(data_normalized) - sequence_length):
    X.append(data_normalized[i:i+sequence_length])
    y.append(data_normalized[i+sequence_length][0])  # Consumo es la primera columna

X = np.array(X)
y = np.array(y)

# Dividir datos en conjunto de entrenamiento y conjunto de prueba
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Definir modelo RNN
model = Sequential([
    LSTM(50, input_shape=(X_train.shape[1], X_train.shape[2])),
    Dense(1)
])

model.compile(optimizer='adam', loss='mse')

# Entrenar modelo
checkpoint_path = f"{sub_estacion}.h5"
checkpoint = ModelCheckpoint(checkpoint_path, monitor='val_loss', save_best_only=True, mode='min', verbose=1)
callbacks_list = [checkpoint]

model.fit(X_train, y_train, epochs=100, batch_size=64, validation_data=(X_test, y_test), callbacks=callbacks_list)

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import load_model

# Cargar datos originales del CSV
data_csv = pd.read_csv('/content/prueba1.csv')

# Cargar el modelo entrenado desde el archivo .h5
model = load_model('ELSALTO.h5')

# Normalizar datos del CSV usando el mismo scaler utilizado durante el entrenamiento
scaler = MinMaxScaler()
data_normalized_csv = scaler.fit_transform(data_csv[['consumo', 'dia', 'mes', 'año', 'hora']])

# Preparar datos para hacer predicciones
X_csv = []
y_csv = []

sequence_length = 24

for i in range(len(data_normalized_csv) - sequence_length):
    X_csv.append(data_normalized_csv[i:i+sequence_length])
    y_csv.append(data_normalized_csv[i+sequence_length][0])

X_csv = np.array(X_csv)
y_csv = np.array(y_csv)

# Realizar predicciones en el conjunto de prueba
predictions = model.predict(X_csv)

# Invertir la normalización en las predicciones y en los datos originales del CSV
predictions_actual = scaler.inverse_transform(np.concatenate((predictions.reshape(-1, 1), X_csv[:, 1:]), axis=1))[:, 0]
y_actual = scaler.inverse_transform(np.concatenate((y_csv.reshape(-1, 1), X_csv[:, 1:]), axis=1))[:, 0]

# Graficar las predicciones junto con los datos reales del CSV
plt.figure(figsize=(10, 6))
plt.plot(predictions_actual, label='Predicciones')
plt.plot(y_actual, label='Datos reales')
plt.xlabel('Tiempo')
plt.ylabel('Consumo')
plt.title('Comparación de Predicciones con Datos Reales')
plt.legend()
plt.show()


```
